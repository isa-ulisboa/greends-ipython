 **Encapsulation** is one of the key concepts in object-oriented programming. Encapsulation refers to the bundling of attributes and methods inside a single class.

 One of the forms of encapsulation is *data hiding*.  Data hiding in Python is typically performed using the `__` *double underscore*  prefix. This makes the class members non-public and isolated from the other classes. More precisely, when you name an attribute or method with two leading underscores, Python automatically renames it by prefixing its name with the class name and a single leading underscore. This renaming process is called *name mangling*.  Suggestion: check this [RealPython article and video](https://realpython.com/python-double-underscore/) on Single and Double Underscores in Python Names.




```python
class Plant:
    def __init__(self,species):
        self.species=species
        self.__price = 10

    def setPrice(self, price):
        self.__price = price

    def __str__(self):
        return f'{self.species} costs {self.__price}'
```

```python
p=Plant('Rose')
print(p)
```

```python
p.__price=20
print(p)
```

```python
p.setPrice(15)
print(p)
```
