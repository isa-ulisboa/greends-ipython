**Polymorphism** is one of the important concepts of object-oriented programming. It simply means more than one form. That is, the same entity (method or operator or object) can perform different operations in different scenarios.

In the example below, there are two classes `Plant`and `Tree` with different attributes. We define the `__str__` method (that is called by `print`) differently in each class.

*Exercise*: define a new method with the same name in each class but that has a distinct functionality.


```python
class Plant:
    def __init__(self,species, price):
        self.species=species
        self.price=price

    def __str__(self):
        return f'{self.species} costs {self.price}'
```

```python
class Tree:
    def __init__(self, species, height_m):
        self.species = species
        self.height_m = height_m

    def __str__(self):
        return f'{self.species} is a tree of {self.height_m} meters.'
```

```python
p=Plant('Rose', 10)
t=Tree('Oak', 30)
print(p)
print(t)
```
